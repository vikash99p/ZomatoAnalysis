**importing libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**reading csv**

```python
df = pd.read_csv('zomato.csv')
df.head()
```

```python
df.shape
```

**cleaning**

```python
df.columns
```

```python
df = df.drop(['url','phone','address','dish_liked','reviews_list'], axis=1)
df.head()
```

```python
df = df.drop(['menu_item'], axis=1)
df.head()
```

```python
df.info()
```

**dropping duplicates**

```python
# drop duplicates
df.drop_duplicates(inplace=True)
df.shape
```

**cleaning Rate column**

```python
# clean the rate column
df['rate'].unique()
```

```python
# remove unique "new","-", and "/5"
def handlerate(value):
    if(value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

**filling null values in rate column with mean**

```python
# filling null values in rate column with mean
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

dropping rest of all null value cause they less

```python
# drop null values
df.dropna(inplace=True)
df.head()
```

renaming columns

```python
# rename columns approx_cost(for two people)	to cost2plates, listed_in(type) to type
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

dropping the listed_in(city) beacuse both are contening same city name and unique value

```python
df = df.drop(['listed_in(city)'], axis=1)
```

```python
df.head()
```

```python
# check unique value in cost2plates
df['cost2plates'].unique()
```

we can see the value having commas which menas value are strincs, commas is risticting the value to becoming a intiger

**removing commas from cost2plate**

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
```

```python
df['cost2plates'] = df['cost2plates'].apply(handlecomma)
```

```python
df['cost2plates'].unique()
```

now every value arre converted in float

```python
df.head()
```

cleaning rest type colummn

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)

```

```python
rest_type
```

all the resturanttypes which have less 1000 resturant no. named as others like resttype

```python
# all the resturant types which have less than 1000 resturant named as rest_types_lessthan1000
rest_types_lessthan1000 = rest_type[rest_type<1000]
rest_types_lessthan1000
```

now the above list shows all the types of resturant which have lees than 1000 resturant

**MAking rest types less than 1000 in freuency as others**
for this im going to write a function

```python
# MAking rest types less than 1000 in freuency as others
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
```

```python
 df['rest_type'] = df['rest_type'].apply(handle_rest_type)
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

**Cleaning loction column**

```python
# value count of loction
df['location'].value_counts()
```

again im going to make cluster of all those loctions having less than300 resturant

```python
# all those location having less 300 resturant named as location_lessthan300
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
location_lessthan300

```

```python
# MAking location less than 300 resturant in freuency as others
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
```

```python
df['location'] = df['location'].apply(handle_location)
```

```python
df['location'].value_counts()
```

```python
df.head()
```
